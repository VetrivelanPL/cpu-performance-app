```python
import pandas as pd

# Load Intel and AMD datasets
df_intel = pd.read_csv('dataset/INTELpartialspecs_adjusted.csv')
df_amd = pd.read_csv('dataset/AMDfullspecs_adjusted.csv')

print("Intel columns:", df_intel.columns.tolist())
print("AMD columns:", df_amd.columns.tolist())
```

```python
# Step 3: Select relevant columns and harmonize names (safe assignments)

intel_cols = ['releaseDate', 'cores', 'threads', 'maxTurboClock', 'baseClock',
              'cache', 'TDP', 'lith', 'manufacturer']

amd_cols = ['launchDate', 'numCores', 'numThreads', 'baseClock', 'maxboostClock',
            'L1Cache', 'L2Cache', 'L3Cache', 'defaultTDP', 'processTech', 'manufacturer']

# Filter columns dynamically based on availability
intel_cols = [col for col in intel_cols if col in df_intel.columns]
amd_cols = [col for col in amd_cols if col in df_amd.columns]

df_intel_selected = df_intel[intel_cols].copy()
df_amd_selected = df_amd[amd_cols].copy()

# Rename AMD columns to Intel schema
amd_rename_map = {
    'launchDate': 'releaseDate',
    'numCores': 'cores',
    'numThreads': 'threads',
    'maxboostClock': 'maxTurboClock',
    'defaultTDP': 'TDP',
    'processTech': 'lith',
}
df_amd_selected = df_amd_selected.rename(columns=amd_rename_map)

# Assign cache columns for Intel: create columns safely
df_intel_selected.loc[:, 'L1Cache'] = None
df_intel_selected.loc[:, 'L3Cache'] = None

if 'cache' in df_intel_selected.columns:
    df_intel_selected.loc[:, 'L2Cache'] = pd.to_numeric(df_intel_selected['cache'], errors='coerce')
    df_intel_selected = df_intel_selected.drop(columns=['cache'])
else:
    df_intel_selected.loc[:, 'L2Cache'] = None

# Convert AMD cache columns to numeric safely
for cache_col in ['L1Cache', 'L2Cache', 'L3Cache']:
    if cache_col in df_amd_selected.columns:
        df_amd_selected[cache_col] = pd.to_numeric(df_amd_selected[cache_col], errors='coerce')

# Hardcode numeric manufacturer: 0 for Intel, 1 for AMD
df_intel_selected.loc[:, 'manufacturer'] = 0
df_amd_selected.loc[:, 'manufacturer'] = 1
```

```python
# Step 4: Merge datasets and handle missing values

import numpy as np

# Concatenate Intel and AMD vertically
df_combined = pd.concat([df_intel_selected, df_amd_selected], ignore_index=True)

# Convert numeric columns to proper dtypes and handle missing values
num_cols = ['cores', 'threads', 'maxTurboClock', 'baseClock', 'L1Cache', 'L2Cache', 'L3Cache', 'TDP']
for col in num_cols:
    if col in df_combined.columns:
        df_combined[col] = pd.to_numeric(df_combined[col], errors='coerce')
        median_val = df_combined[col].median()
        df_combined[col] = df_combined[col].fillna(median_val)

# Handle categorical columns
cat_cols = ['lith']
for col in cat_cols:
    if col in df_combined.columns:
        df_combined[col] = df_combined[col].fillna('Unknown')

# Create totalCache feature
df_combined['totalCache'] = df_combined[['L1Cache', 'L2Cache', 'L3Cache']].sum(axis=1, min_count=1).fillna(0)

# Create coreClock as cores * baseClock
df_combined['coreClock'] = df_combined['cores'] * df_combined['baseClock']

# Drop irrelevant or redundant columns if any remain
drop_cols = ['releaseDate', 'lith']  # You may encode lith later if needed
df_combined.drop(columns=drop_cols, inplace=True, errors='ignore')

# Create performance_score target as weighted sum of key features
df_combined['performance_score'] = (
    df_combined['cores'] * df_combined['maxTurboClock'] * 0.4 +
    df_combined['threads'] * 0.2 +
    df_combined['totalCache'] * 0.3 -
    df_combined['TDP'] * 0.1
)

print("Combined dataset shape:", df_combined.shape)
print(df_combined.head())
```

```python
# Define performance tiers based on quantiles (33% and 66%)
bins = [float('-inf'),
        df_combined['performance_score'].quantile(0.33),
        df_combined['performance_score'].quantile(0.66),
        float('inf')]
labels = ['Low', 'Medium', 'High']

df_combined['performance_tier'] = pd.cut(df_combined['performance_score'], bins=bins, labels=labels)

# Encode tier labels numerically for ML
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_combined['performance_tier_encoded'] = le.fit_transform(df_combined['performance_tier'])

print(df_combined[['performance_score', 'performance_tier', 'performance_tier_encoded']].head())

# Feature matrix and targets for regression and classification
X = df_combined.drop(columns=['performance_score', 'performance_tier', 'performance_tier_encoded'])
y_reg = df_combined['performance_score']
y_clf = df_combined['performance_tier_encoded']

# Split data into train/test sets
from sklearn.model_selection import train_test_split

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.2, random_state=42)
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X, y_clf, test_size=0.2, random_state=42)

print("Regression train shape:", X_train_reg.shape)
print("Classification train shape:", X_train_clf.shape)
```

```python
import warnings

# Suppress specific categories of warnings
warnings.filterwarnings('ignore', category=UserWarning)
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
import xgboost as xgb
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import numpy as np

# Impute missing values in features
imputer = SimpleImputer(strategy='mean')
X_train_reg_imputed = imputer.fit_transform(X_train_reg)
X_train_clf_imputed = imputer.fit_transform(X_train_clf)

# Define regressors
regressors = {
    'LinearRegression': LinearRegression(),
    'RandomForest': RandomForestRegressor(random_state=42),
    'XGBoost': xgb.XGBRegressor(random_state=42, eval_metric='rmse')
}

print("Regression model cross-validation R² scores:")
for name, model in regressors.items():
    scores = cross_val_score(model, X_train_reg_imputed, y_train_reg, cv=5, scoring='r2')
    print(f"{name}: mean R²={np.mean(scores):.4f} ± std={np.std(scores):.4f}")

# Define classifiers with scaling pipeline
classifiers = {
    'LogisticRegression': make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000, random_state=42)),
    'RandomForest': RandomForestClassifier(random_state=42),
    'XGBoost': xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss')
}

print("\nClassification model cross-validation accuracies:")
for name, model in classifiers.items():
    scores = cross_val_score(model, X_train_clf_imputed, y_train_clf, cv=5, scoring='accuracy')
    print(f"{name}: mean accuracy={np.mean(scores):.4f} ± std={np.std(scores):.4f}")
```

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import RandomizedSearchCV
import numpy as np
import xgboost as xgb
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Define parameter grids for hyperparameter search
rf_reg_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

xgb_reg_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 1.0]
}

rf_clf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

lr_param_grid = {
    'logisticregression__C': [0.01, 0.1, 1, 10, 100]
}

xgb_clf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 1.0]
}

# Impute missing values
imputer = SimpleImputer(strategy='mean')
X_imputed_reg = imputer.fit_transform(X_train_reg)
X_imputed_clf = imputer.fit_transform(X_train_clf)

# Create base models
rf_reg = RandomForestRegressor(random_state=42)
xgb_reg = xgb.XGBRegressor(random_state=42, eval_metric='rmse')

rf_clf = RandomForestClassifier(random_state=42)
lr_pipe = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000, random_state=42))
xgb_clf = xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss')

# Hyperparameter tuning - Regression
print("Starting hyperparameter tuning for regression models...")

rf_reg_search = RandomizedSearchCV(rf_reg, rf_reg_param_grid, n_iter=10, cv=3, scoring='r2', random_state=42, n_jobs=-1)
rf_reg_search.fit(X_imputed_reg, y_train_reg)
print("Best RF regression params:", rf_reg_search.best_params_)
print("Best RF regression R²:", rf_reg_search.best_score_)

xgb_reg_search = RandomizedSearchCV(xgb_reg, xgb_reg_param_grid, n_iter=10, cv=3, scoring='r2', random_state=42, n_jobs=-1)
xgb_reg_search.fit(X_imputed_reg, y_train_reg)
print("Best XGB regression params:", xgb_reg_search.best_params_)
print("Best XGB regression R²:", xgb_reg_search.best_score_)

# Hyperparameter tuning - Classification
print("\nStarting hyperparameter tuning for classification models...")

rf_clf_search = RandomizedSearchCV(rf_clf, rf_clf_param_grid, n_iter=10, cv=3, scoring='accuracy', random_state=42, n_jobs=-1)
rf_clf_search.fit(X_imputed_clf, y_train_clf)
print("Best RF classification params:", rf_clf_search.best_params_)
print("Best RF classification accuracy:", rf_clf_search.best_score_)

lr_search = RandomizedSearchCV(lr_pipe, lr_param_grid, n_iter=5, cv=3, scoring='accuracy', random_state=42, n_jobs=-1)
lr_search.fit(X_imputed_clf, y_train_clf)
print("Best Logistic Regression params:", lr_search.best_params_)
print("Best Logistic Regression accuracy:", lr_search.best_score_)

xgb_clf_search = RandomizedSearchCV(xgb_clf, xgb_clf_param_grid, n_iter=10, cv=3, scoring='accuracy', random_state=42, n_jobs=-1)
xgb_clf_search.fit(X_imputed_clf, y_train_clf)
print("Best XGB classification params:", xgb_clf_search.best_params_)
print("Best XGB classification accuracy:", xgb_clf_search.best_score_)
```

```python
import joblib
from sklearn.impute import SimpleImputer
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, classification_report, confusion_matrix
import xgboost as xgb
import numpy as np

# Impute training and test data
imputer = SimpleImputer(strategy='mean')
X_train_reg_imputed = imputer.fit_transform(X_train_reg)
X_test_reg_imputed = imputer.transform(X_test_reg)

X_train_clf_imputed = imputer.fit_transform(X_train_clf)
X_test_clf_imputed = imputer.transform(X_test_clf)

# Define final models with tuned hyperparameters
final_xgb_reg = xgb.XGBRegressor(
    subsample=0.7,
    n_estimators=100,
    max_depth=6,
    learning_rate=0.2,
    random_state=42,
    eval_metric='rmse'
)

final_xgb_clf = xgb.XGBClassifier(
    subsample=0.7,
    n_estimators=200,
    max_depth=9,
    learning_rate=0.2,
    random_state=42,
    use_label_encoder=False,
    eval_metric='mlogloss'
)

# Train final models
final_xgb_reg.fit(X_train_reg_imputed, y_train_reg)
final_xgb_clf.fit(X_train_clf_imputed, y_train_clf)

# Evaluate regression model
y_pred_reg = final_xgb_reg.predict(X_test_reg_imputed)
print("XGBoost Regression Evaluation:")
print("R²:", r2_score(y_test_reg, y_pred_reg))
rmse = np.sqrt(mean_squared_error(y_test_reg, y_pred_reg))
print("RMSE:", rmse)

# Evaluate classification model
y_pred_clf = final_xgb_clf.predict(X_test_clf_imputed)
print("\nXGBoost Classification Evaluation:")
print("Accuracy:", accuracy_score(y_test_clf, y_pred_clf))
print("Classification Report:\n", classification_report(y_test_clf, y_pred_clf))
print("Confusion Matrix:\n", confusion_matrix(y_test_clf, y_pred_clf))

# Save models and imputer
joblib.dump(final_xgb_reg, 'final_xgb_regressor.joblib')
joblib.dump(final_xgb_clf, 'final_xgb_classifier.joblib')
joblib.dump(imputer, 'imputer.joblib')

print("\nFinal models and imputer saved successfully.")
```
